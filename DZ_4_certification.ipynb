<font size=4.5 color=#3a75c4>**2.2. Изучите распределение этажей домов.**</font>
```
1. Постройте график;
2. Сделайте выводы.
```

````python
# 1 ВАРИАНТ. КРУГОВАЯ ДИАГРАММА
plt.figure(figsize=(9, 9)).set_facecolor('#E5E4E2')
data_1 = df_1["floors"]
data_1 = data_1.value_counts()
explode = (0.01, 0.03, 0.05, 0.10, 0.15, 0.2)
grouped = df_1.groupby('floors')
plt.title('Распределение домов в зависимости от этажей', color = '#3a75c4')
plt.pie(data_1.values, explode = explode, autopct='%.1f%%', radius=1.1, pctdistance=0.75, labels=data_1.index, wedgeprops={'alpha':1})
# fig.set_facecolor('lightgrey')
plt.legend(grouped.groups.keys());
````

````python
# 2 ВАРИАНТ. ВЕРТИКАЛЬНАЯ ДИАГРАММА
data_1 = df_1["floors"].value_counts()
Kol_home=df_1["floors"].value_counts().sum() #Количество домов должно быть равно количеству строк (21613 rows)
# Рассчитываем количество домов в процентном соотношении, в зависимости от этажности
kol_home_procent = []
for i in range(data_1.shape[0]):  # .shape - получаем количество групп, разделенных в зависимости от этажей.
    procent = (data_1.values[i] / Kol_home) * 100 
    kol_home_procent.append(round(procent,2))
# Строим dертикальную диаграмму
plt.figure(figsize=(9,9))
diagrama=plt.bar(data_1.index,data_1.values, color = ["#c43aba","#3a75c4","#3ac444","#c4893a","#c4443a","#dd214c"])
plt.title('Распределение домов в зависимости от этажности', color = '#3a75c4')
plt.xlabel("Количество этажей")
plt.ylabel("Количество домов")
plt.grid(color='#B5B8B1', alpha=0.30)
# "Рисуем" проценты над столбиками диаграммы
i = 0
for p in diagrama:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
             y+height*1.015,
             str(kol_home_procent[i])+'%',
             ha='center',
             weight='bold')
    i+=1
plt.show()
````
